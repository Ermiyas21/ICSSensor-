```python
import matplotlib.pyplot as plt

x_axis = ['1', '5', '10', '15', '20','25','30']
y_axis = ['0.47','0.54', '0.27','0.41','0.45','0.75','0.49']

colors = ['green', 'blue', 'purple', 'brown', 'teal']
plt.bar(x_axis, y_axis, color=colors)
plt.title('RMSE Performance for SWaT dataset ', fontsize=14)
plt.xlabel('resample rate', fontsize=14)
plt.ylabel('RMSE', fontsize=14)
#plt.grid(True)
plt.show()
```
