```python

import numpy as np
import matplotlib.pyplot as plt

X = ['RMSE','Precsion','Recall']
Ygirls = [0.160,0.87,0.470]
Zboys = [0.306,0.46,0.970]

X_axis = np.arange(len(X))

plt.bar(X_axis - 0.2, Ygirls, 0.4, label = 'AE')
plt.bar(X_axis + 0.2, Zboys, 0.4, label = 'Resample AE')

plt.xticks(X_axis, X)
plt.xlabel("performance metrics")
plt.ylabel("performance (%)")
plt.title("Performance comparison for HAI dataset")
plt.legend()
plt.show()
#plt.savefig('Perfomance for HAI.png')
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 7
ind = np.arange(N)
width = 0.25

xvals = [0.47,1.21,0.45, 0.62,0.43,0.66,0.42] # RMSE
bar1 = plt.bar(ind, xvals, width, color = 'r')

yvals = [0.35,0.46,0.59,0.98,0.78,0.57,0.61] # pre
bar2 = plt.bar(ind+width, yvals, width, color='g')

zvals = [0.44,0.82,0.36,0.63,0.92,0.31,0.37] #recall
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')

plt.xlabel("Dates")
plt.ylabel('Scores')
plt.title("Players Score")

#plt.xticks(ind+width,['RMSE', 'Pre', 'reca'])
plt.legend( (bar1, bar2, bar3), ('raw data', '5 second', '10 second','15 second','15 second','20 second','25 second','30 second') )
plt.show()
```
